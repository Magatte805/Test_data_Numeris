# Note de synthèse

## Détection d'opportunités de missions de conseil à partir d'un Fichier des Écritures Comptables (FEC)

## 1. Contexte

L'objectif de ce projet était d'exploiter un Fichier des Écritures Comptables (FEC) anonymisé afin d'identifier automatiquement des opportunités de missions de conseil pour un cabinet d'expertise comptable.

L'enjeu consistait à transformer un volume important d'écritures comptables en informations directement exploitables par les associés afin de les aider à identifier les entreprises à accompagner en priorité.

---

## 2. Démarche

Notre démarche s'est articulée autour de quatre étapes.

### Étape 1 – Exploration des données

Nous avons commencé par explorer les deux jeux de données fournis : le FEC anonymisé et le référentiel clients.

Cette étape a permis de comprendre la structure des données, d'identifier les principales colonnes, de vérifier la qualité des données et de préparer les traitements suivants. Compte tenu du volume du FEC, les données ont également été converties au format Parquet afin d'optimiser les performances de lecture.

### Étape 2 – Construction des indicateurs

À partir des écritures comptables, nous avons construit plusieurs indicateurs financiers par entreprise :

* chiffre d'affaires mensuel et annuel ;
* position de trésorerie ;
* créances clients ;
* dettes fournisseurs ;
* Besoin en Fonds de Roulement (BFR) ;
* DSO (Days Sales Outstanding).

Ces indicateurs constituent la base de l'analyse métier réalisée dans les étapes suivantes.

### Étape 3 – Identification des opportunités de missions

Les indicateurs calculés ont été utilisés afin de détecter automatiquement des opportunités de missions de conseil.

Nous avons défini six missions pertinentes :

* Optimisation du recouvrement clients ;
* Optimisation du BFR ;
* Pilotage de la trésorerie ;
* Accompagnement au pilotage financier ;
* Optimisation de la structure financière ;
* Diagnostic financier approfondi.

Chaque mission est déclenchée par plusieurs signaux financiers cohérents et est accompagnée d'un niveau de priorité ainsi que d'une explication destinée aux associés.

### Étape 4 – Restitution

Les résultats sont restitués sous la forme d'un dashboard interactif développé avec Apache ECharts (via la bibliothèque Python **pyecharts**).

Le dashboard offre une double lecture :

* une vision globale du portefeuille grâce à plusieurs visualisations ;
* un tableau des opportunités permettant d'identifier rapidement les clients à contacter, les missions recommandées et les éléments financiers ayant conduit à ces recommandations.

---

## 3. Hypothèses retenues

Certaines informations n'étant pas directement disponibles dans le FEC, plusieurs hypothèses ont été retenues et documentées dans le notebook.

Les principales hypothèses sont les suivantes :

* le chiffre d'affaires est estimé à partir du solde des comptes de produits (classe 7) ;
* la trésorerie est calculée à partir des principaux comptes de disponibilités (banques, caisse et comptes assimilés) ;
* les créances clients et les dettes fournisseurs sont estimées à partir des comptes comptables correspondants ;
* le DSO est calculé à partir des créances clients et du chiffre d'affaires annuel estimé ;
* les seuils de détection des missions reposent sur les quartiles observés dans le portefeuille plutôt que sur des valeurs fixes.

Ces hypothèses permettent d'appliquer une méthode homogène à l'ensemble des entreprises étudiées.

---

## 4. Limites

Cette approche présente plusieurs limites.

Le FEC ne contient pas directement tous les indicateurs financiers recherchés. Certains calculs reposent donc sur des approximations comptables, clairement explicitées dans le notebook.

Par ailleurs, les règles de détection des missions reposent sur une logique métier volontairement simple afin de garantir leur interprétabilité. Elles constituent une aide à la décision et ne remplacent pas l'analyse réalisée par un expert-comptable.

Enfin, le projet repose sur un seul exercice comptable et ne permet pas d'étudier l'évolution des entreprises dans le temps.

---

## 5. Pistes d'amélioration

Plusieurs améliorations pourraient être envisagées :

* intégrer plusieurs exercices comptables afin d'analyser les tendances ;
* enrichir les données avec des informations externes (secteur d'activité, effectifs, taille de l'entreprise, etc.) ;
* développer un score de priorité plus élaboré en combinant plusieurs indicateurs ;
* ajouter de nouveaux indicateurs financiers afin d'affiner les recommandations ;
* améliorer le dashboard avec davantage d'interactivité et de fonctionnalités d'exploration.

---

## Conclusion

Ce projet montre qu'il est possible de transformer un Fichier des Écritures Comptables en un véritable outil d'aide à la décision.

En combinant traitement de données, indicateurs financiers, règles métier et visualisation interactive, il devient possible d'identifier rapidement les entreprises présentant un potentiel de missions de conseil et de fournir aux associés une restitution claire, exploitable et directement orientée vers l'action.
